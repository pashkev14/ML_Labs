```python
table = pd.read_csv('/content/global-weather-repository/GlobalWeatherRepository.csv')
table.describe()
```

С помощью метода `DataFrame.describe()` получили характеристику по каждому числовому атрибуту. Значения среднего и СКО представлены в строках mean и std, соответственно.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
features = table.columns
categorical_features = [features[0], features[1], features[4], features[9], features[13], features[-2]]
datetime_features = [features[5], features[6], features[35], features[36], features[37], features[38]]
numerical_features = [x for x in features if x not in categorical_features and x not in datetime_features]
```

```python
for feature in numerical_features:
  plt.figure(figsize=(8, 6))
  plt.hist(table[feature], bins=30, alpha=0.7, color='yellow', edgecolor='black')
  plt.title(f'Гистограмма {feature}')
  plt.xlabel(feature)
  plt.ylabel('Частота')
  plt.grid(axis='y', alpha=0.75)
  plt.show()
```

Построили гистограммы для числовых признаков.

```python
def count_top_50(feature):
  counts = table[feature].value_counts()
  counts = counts.sort_values(ascending=False)
  if len(counts) > 50:
    counts_final = counts[0:50]
    counts_final['other'] = counts[50:].sum()
    return counts_final
  else:
    return counts
```

```python
for feature in categorical_features:
  plt.figure(figsize=(10, 6))
  counted_values = count_top_50(feature)
  if not counted_values.empty:
    counted_values.plot(kind='bar', width=0.8, color='yellow', edgecolor='black')
    plt.title(f'Гистограмма {feature}', fontsize=16)
    plt.xlabel(feature, fontsize=14)
    plt.ylabel('Частота', fontsize=14)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', alpha=0.75)
    plt.tight_layout()
    plt.show()
  else:
    print(f'Нет данных для {feature}')
```

Построили гистограммы для категориальных признаков.

```python
def count_top_50(feature):
  counts = table[feature].value_counts()
  counts = counts.sort_values(ascending=False)
  if len(counts) > 50:
    counts_final = counts[0:50]
    return counts_final
  else:
    return counts
```

```python
for feature in categorical_features:
  plt.figure(figsize=(10, 6))
  counted_values = count_top_50(feature)
  if not counted_values.empty:
    counted_values.plot(kind='bar', width=0.8, color='yellow', edgecolor='black')
    plt.title(f'Гистограмма {feature}', fontsize=16)
    plt.xlabel(feature, fontsize=14)
    plt.ylabel('Частота', fontsize=14)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', alpha=0.75)
    plt.tight_layout()
    plt.show()
  else:
    print(f'Нет данных для {feature}')
```

Определили, что пропусков в датасете нет.

```python
def code_categorical_features_to_numerical(dataframe, column):

    def convert_to_index(value, uniq_values):
        return uniq_values.index(value)

    lists_uniq_values = sorted(list(set(dataframe[column])))
    dataframe[column] = dataframe[column].apply(lambda x: convert_to_index(x, lists_uniq_values))

for feature in categorical_features:
    code_categorical_features_to_numerical(table, feature)

table.head()
```

Закодировали категориальные признаки в числовые.

```python
plt.figure(figsize=(30,20), dpi=80)
sns.heatmap(table[numerical_features + categorical_features].corr(), xticklabels=table[numerical_features + categorical_features].corr().columns, yticklabels=table[numerical_features + categorical_features].corr().columns, center=0, annot=True)

plt.title('Диаграмма корреляций признаков', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
