```python
import pandas as pd
```

```python
table = pd.read_csv('/content/global-weather-repository/GlobalWeatherRepository.csv')
table.describe()
```

–° –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ `DataFrame.describe()` –ø–æ–ª—É—á–∏–ª–∏ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫—É –ø–æ –∫–∞–∂–¥–æ–º—É —á–∏—Å–ª–æ–≤–æ–º—É –∞—Ç—Ä–∏–±—É—Ç—É. –ó–Ω–∞—á–µ–Ω–∏—è —Å—Ä–µ–¥–Ω–µ–≥–æ –∏ –°–ö–û –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω—ã –≤ —Å—Ç—Ä–æ–∫–∞—Ö mean –∏ std, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
features = table.columns
categorical_features = [features[0], features[1], features[4], features[9], features[13], features[-2]]
datetime_features = [features[5], features[6], features[35], features[36], features[37], features[38]]
numerical_features = [x for x in features if x not in categorical_features and x not in datetime_features]
```

```python
for feature in numerical_features:
  plt.figure(figsize=(8, 6))
  plt.hist(table[feature], bins=30, alpha=0.7, color='yellow', edgecolor='black')
  plt.title(f'–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ {feature}')
  plt.xlabel(feature)
  plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
  plt.grid(axis='y', alpha=0.75)
  plt.show()
```

–ü–æ—Å—Ç—Ä–æ–∏–ª–∏ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –¥–ª—è —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.

```python
def count_top_10(feature):
  counts = table[feature].value_counts()
  counts = counts.sort_values(ascending=False)
  if len(counts) > 20:
    counts_final = counts[0:20]
    counts_final['other'] = counts[20:].sum()
    return counts_final
  else:
    return counts

count_top_10('country')
```

```python
for feature in categorical_features:
  plt.figure(figsize=(10, 6))
  counted_values = count_top_10(feature)
  if not counts.empty:
    counted_values.plot(kind='bar', width=0.8, color='yellow', edgecolor='black')
    plt.title(f'–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ {feature}', fontsize=16)
    plt.xlabel(feature, fontsize=14)
    plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞', fontsize=14)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', alpha=0.75)
    plt.tight_layout()
    plt.show()
  else:
    print(f'–ù–µ—Ç –¥–∞–Ω–Ω—ã—Ö –¥–ª—è {feature}')
```
